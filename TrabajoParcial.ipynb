```python
%%writefile Tparcial.cpp
#include <iostream>
#include <functional>
#include <fstream>
#include <sstream>
#include <string>
using namespace std;
template<typename T>
class Parcial{
    int cant;
    int size;
    T* arr;
public:
    Parcial(){size=0; cant=10; arr=new T[cant];}
    void incrementa(){
        cant*=1.5;
        T* tmp=new T[cant];
        for(size_t i=0;i<size;++i){
            tmp[i]=arr[i];
        }delete arr;
        arr=tmp;
    }
    void insertar(T e){
        if(size==cant) incrementa();
        arr[size++]=e;
    }
    void show(function<void(T)> impresion){
        for(size_t i=0; i<size; i++){
            impresion(arr[i]);
        }
        cout<<endl;
    }    
};
int main(){
    Parcial<string> P;
    string cad; string sec;
    ifstream archivo("ejm.csv");
    
    auto lectura=[](string ca,string seg,char c){        
        stringstream lec(ca);
        getline(lec,seg,c);
        return seg;                
    };
    while(getline(archivo,cad)){
        lectura(cad,sec,',');
        P.insertar(sec);        
    }
    P.show([](string s){cout<<s<<" ";});
    
}
```

```python
%%writefile TP.cpp

#include <iostream>
#include <functional>
#include <sstream>
#include <fstream>
#include <vector>
#include <string>

using namespace std;
template <typename Tipo>
struct Nodo{//Nodo Doble
      Tipo elemento;
      Nodo* siguiente;//puntero al siguiente nodo
      Nodo* anterior;
      Nodo(Tipo e,Nodo* s=nullptr,Nodo* a=nullptr ):elemento(e),siguiente(s),anterior(a){}
};

template <class Tipo>
class Lista{
    Nodo<Tipo>* inicio;
    Nodo<Tipo>* fin;
    long cantidad;

    public:
    Lista(){
        inicio=fin=nullptr;
        cantidad=0;
    }
    void push_front(Tipo valor){
        Nodo<Tipo>* nuevo=new Nodo<Tipo>(valor);//3
        if(cantidad==0)//1+(2,5)
           inicio=fin=nuevo;//2
        else {
            inicio->anterior=nuevo;//2
            nuevo->siguiente=inicio;//2
            inicio=nuevo;//1
            }
        ++cantidad;//2
    }
    //Analisis detallado: 8
    //Big 0: O(1)
    void push_back(Tipo valor) {
        Nodo<Tipo>* nuevo = new Nodo<Tipo>(valor);//2
        if(cantidad==0) {//1+(2,5)
            inicio = fin = nuevo;
        } else {
            fin->siguiente = nuevo;//2
            nuevo->anterior=fin;//
            fin = nuevo;//1
        }
        ++cantidad;//2
    }
    //Analisis detallado: 6
    //Big O: O(1) 

    void Push_position(Tipo valor, long pos){
        if(pos<=0)
          push_front(valor);
        else if (pos>=cantidad)
          push_back(valor);
        else{
            Nodo<Tipo>* aux=inicio;
            Nodo<Tipo>* nuevo=new Nodo<Tipo>(valor);
             for(int i=0;i<pos-1;++i)
                aux=aux->siguiente;
            nuevo->siguiente=aux->siguiente;
            nuevo->anterior=aux;
            aux->siguiente=nuevo;
            nuevo->siguiente->anterior=nuevo;
            ++cantidad;
        }
    }
    Tipo recuperarPosicion(int pos) {
			if (pos <= cantidad) {
				Nodo<Tipo>* aux = inicio;
				for (int i=0;i<pos;i++){
            aux=aux->siguiente;
        }
				//cout << "\nEl elemento en la posicion "<< pos << " es: " << aux->elemento<<endl;
        return aux->elemento;			
      }
		}
    bool buscar(Tipo e, function <bool(Tipo, string)> Com){
        Nodo<Tipo>* aux=inicio;
        int c=0;
        while(aux != nullptr)
        {
            if(Com(aux->elemento, e)){             
              cout << "Elemento encontrado: " << aux->elemento<<" posicion: "<<c<<endl;
               return true; 
            } 
            c++;
            aux=aux->siguiente;
        }
        return false;
     }
    void mostrar(function<void(Tipo)> funcion_Mostrar){
        Nodo<Tipo>* aux=inicio;
        while(aux->siguiente!=nullptr){
            funcion_Mostrar(aux->elemento);
            aux=aux->siguiente;
          }
          funcion_Mostrar(aux->elemento);
    }
   long size(){
       return cantidad;
   } 
};

template  <class T>
class Arreglo{
    int cant;
    int size;
    T* arr;
public:
    Arreglo(){size=0; cant=10; arr=new T[cant];}
    void incrementa(){
        cant*=1.5;//2
        T* tmp=new T[cant];//3
        for(size_t i=0;i<size;++i){//1+1+n(1+...+2)
            tmp[i]=arr[i];//3
        }delete arr;//1
        arr=tmp;//1
    }
    //Analisis detallado: 8+7n
    //Big O:  O(n)
    void push_back(T e){
        if(size==cant) incrementa();
        arr[size++]=e;
    }
    void show(function<void(T)> impresion){
        for(size_t i=0; i<size; i++){//2+n(1+..+2)
            impresion(arr[i]);//2
        }
        cout<<endl;//1
    } 
    int Tamanio(){return size;}   //1
};
//Analisis detallado: 4+5n
//Big(o): O(n)


class Usuario1{
    private:
    int id;
    string nombre;
    string apellido;
    int edad;
    string codigo;

    public:    
    Usuario1(int id=0,string nombre="",string apellido="", int edad=0, string codigo=""){
      this->id=id;
      this->nombre=nombre;
      this->apellido=apellido;
      this->edad=edad;
      this->codigo=codigo;

    }
    int getId(){return this->id;}
    string getNombre(){return nombre;}
    string getApellido(){return apellido;}
    int getEdad(){return edad;}
    string getFecha(){return codigo;}
    friend ostream& operator<<(ostream& o, Usuario1 & v){
        o<<v.id<<" "<<v.nombre<<" "<<v.apellido<<" "<<v.edad<<" "<<v.codigo<<endl;
        return o;
    }
    friend bool operator<=(Usuario1 & a,Usuario1 & b){
        return a.edad<b.edad;        
    }
};

class Usuario2{
    string comentarios;
    public:
    Usuario2(string comentarios=""){
        this->comentarios=comentarios;
    }
    string getComentarios(){return comentarios;}

};

class ADMINISTRADOR{
  Lista<Usuario1> usr;
  Arreglo<Usuario2> usr2;
public:
      ADMINISTRADOR(){ }  
      void printing() {          
         usr.mostrar([](Usuario1 usr){cout<<usr<<" ";}); 
         cout << endl;
      }

  void lectura(){
    int id;
    string nombre;
    string apellido;
    int edad;
    string codigo;         
    string temp;
      ifstream archivo("archivo.txt");
      getline(archivo,temp);
      while(archivo>>id>>nombre>>apellido>>edad>>codigo){
          usr.push_back(Usuario1(id,nombre, apellido,edad,codigo));
      }
      cout<<"Total: "<<usr.size()<<endl;
   
  }
    void salir(Lista<Usuario1> p,function<void(Lista<Usuario1>)> criterio){
    criterio(p);
    //archivo.close();
}

  void escritura(){ 
    ofstream archi("Administrador.txt");
    archi<<endl;
    string op;

    do{
        cout<<"C:/usr1>";
        cin>>op;

       if(op=="Menu"){
           cout<<"     Menu      "<<endl;
           cout<<"----------------"<<endl;
           cout<<"1. CrearArchivos "<<endl;
           cout<<"2. CrearUsuario "<<endl;
           cout<<"3. Buscar por"<<endl;
           cout<<"4. Mostrar"<<endl;
           cout<<"5. Exit"<<endl;
       }
       if (op=="1" || op=="CrearArchivos"){
           string arq;
           cin>>arq;
           ofstream Archi1(arq);
           archi<<arq<<endl;
           cout<<" Archivo creado correctamente"<<endl;
           Archi1.close();
       }
       if (op=="2" || op=="CrearUsuario"){
           cout<<"1: usuario 1"<<endl;
           cout<<"2. usuario 2"<<endl;
           cin>>op;
           if(op=="1"){
            int id;
            string nombre;
            string apellido;
            int edad;
            string codigo;  
           string comentarios;
           cout<<"Id: ";cin>>id;
           cout<<"NOmbre: ";cin>>nombre;
           cout<<"apellido: ";cin>>apellido;
           cout<<"edad: ";cin>>edad;
           cout<<"codigo: ";cin>>codigo;
           usr.push_back(Usuario1(id,nombre, apellido,edad,codigo));
           //usr2.push_back(Usuario2(comentarios));
           archi<<comentarios;
           archi<<id<<"|"<<nombre<<"|"<<apellido<<"|"<<edad<<"|"<<codigo<<endl;
           archi.close();

           }
           if(op=="2"){ 
               string comentario;
               cout<<"Comenta: ";cin>>comentario;
               usr2.push_back(Usuario2(comentario));
               ofstream Bli("test01.txt");
               Bli<<comentario<<endl;
               for(int i=0;i<usr2.Tamanio();i++){
                   usr2.push_back(Usuario2(comentario));
               }

           }         
       }

       if (op=="3" || op=="busqueda"){
           int caso;
           cout<<"1. Igual a: "<<endl;
           cout<<"2. Inicia con: "<<endl;
           cout<<"3. finaliza con: "<<endl;
           cout<<"4. Esta contenido en: "<<endl;
           cout<<"5. NO esta contenido en: "<<endl;
           cout<<"..."<<endl;
           cin>>caso; 
           if(caso==1){
               string st1=" ";
               cout<<"Escribir nombre: "<<endl; 
               cin>> st1;
               for(int i=0; i<usr.size();i++){
                   Usuario1 u = usr.recuperarPosicion(i);
                   if(u.getNombre()==st1){
                       cout<<u.getNombre();
                   }
               }
           }
           if(caso==2){
                  char li=' ';
               cout<<"Primer letra: "<<endl; 
               cin>> li;
               for(int i=0; i<usr.size();i++){
                   Usuario1 u = usr.recuperarPosicion(i);
                   if(u.getNombre()[0]==li){
                       cout<<u.getNombre();
                   }
               }
            }
            if(caso==3){
                  char lf=' ';
               cout<<"Ãltima letra: "<<endl; 
               cin>> lf;
               for(int i=0; i<usr.size();i++){
                   Usuario1 u = usr.recuperarPosicion(i);
                   if(u.getNombre()[u.getNombre().length()]==lf){
                       cout<<u.getNombre();
                   }
               }
            }
            if(caso==4){ 
               char ex=' '; cout<<"Buscar si contiene la letra: "<<endl; 
               cin>> ex;
               for(int i=0; i<usr.size();i++){
                   Usuario1 u = usr.recuperarPosicion(i);
                   for(int j=0; j<u.getNombre().length(); j++)
                      if(u.getNombre()[j]==ex){
                          cout<<u.getNombre();
                      }
               }
              }
              if(caso==5){
               char ex=' '; cout<<"Buscar si no contiene la letra: "<<endl; 
               cin>> ex; 
               for(int i=0; i<usr.size();i++){
                   Usuario1 u = usr.recuperarPosicion(i);
                   for(int j=0; j<u.getNombre().length(); j++)
                      if(u.getNombre()[j]!=ex){
                          cout<<u.getNombre();
                      }
               }
           }           
       }

       if(op=="Mostrar" || op=="4"){ 
           printing();
       }
   
    }while(op !="Exit" || op!="5");
  }
};

int main(){

    ADMINISTRADOR ur;
    ur.lectura();
    //ur.printing();
    ur.escritura();

  return 0;
}
```
